### Temporal Analysis: Loan Applications by Month and Purpose

```python
# Create a new column called ApplicationMonth
bank_loans['ApplicationDate']=pd.to_datetime(bank_loans['ApplicationDate'])

bank_loans['ApplicationMonth']=bank_loans['ApplicationDate'].dt.month

bank_loans
```

```python
# Determine the number of applications in each month
bank_loans.groupby('ApplicationMonth')['CustomerNo'].count()
```

```python
# Determine the number of applications for each loan purpose in each month
bank_loans.groupby(['ApplicationMonth','LoanPurpose'])['CustomerNo'].count()
```

```python
# Determine the number of different loan purposes in each month
bank_loans.groupby('ApplicationMonth')['LoanPurpose'].nunique()
```

### Business Rule Check: Identifying Censored Loan Purposes

```python
# Create a function called check_cat_censored
def check_cat_censored(loan_purpose):
    censored_keywords=['Car','Appliance']
    for keyword in censored_keywords:
        if keyword in loan_purpose:
            return True
    else:
        return False
```

```python
# Create a new column called requires_censored
bank_loans['requires_censored']=bank_loans['LoanPurpose'].apply(check_cat_censored)

bank_loans
```

```python
# Identify customers that satisfy the two conditions at the same time
filtered_customers=bank_loans.loc[(bank_loans['Years']>=3) & bank_loans['requires_censored']==True]
filtered_customers
```

### Data Integration: Merging Loan Data with Monthly Information (XML)

```python
# Load the XML file's data into a DataFrame named mon_info
mon_info=pd.read_xml("/Users/rohanchawla/Downloads/bank_extra.xml")
mon_info
```

```python
# Combine bank_loans and mon_info into one single DataFrame loans
mon_info['Month_Number']=pd.to_datetime(mon_info['Mon'],format='%b').dt.month

loans=pd.merge(bank_loans,mon_info, left_on='ApplicationMonth', right_on='Month_Number')

loans
```

```python
# Sort loans by the column ApplicationMonth descendingly first and then by the customer number ascendingly. Then, save the sorted results in a new DataFrame sorted_loans.
sorted_loans=loans.sort_values(by=['ApplicationMonth','CustomerNo'],ascending=[False,True]).reset_index(drop=True)

sorted_loans
```

### Data Integration: Merging Loan Data with JSON Information

```python
# Load the JSON file's data into a DataFrame named loan_info
loan_info=pd.read_json('/Users/rohanchawla/Downloads/bank_extra.json')

loan_info
```

```python
# Combine sorted_loans and loan_info into one single DataFrame loans_final
loan_info_t = loan_info.T.reset_index()
loan_info_t

# Merge with sorted_loans
loans_final=pd.merge(sorted_loans,loan_info_t,left_on='LoanPurpose',right_on='index')
loans_final
```

```python
# Sort loans_final by the customer number descendingly and save the sorted results in a new DataFrame sorted_loans_final
sorted_loans_final=loans_final.sort_values(by='CustomerNo',ascending=False).reset_index(drop=True)
sorted_loans_final
```
